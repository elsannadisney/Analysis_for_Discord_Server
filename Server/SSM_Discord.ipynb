```python
from konlpy.tag import Okt
import re
with open('파일링크.txt', 'r', encoding='utf-8') as f:
  content = f.read()
p = re.compile("[\Wa-zA-Z0-9_]+")
content = re.sub(p, "", content)
okt = Okt()
morphs = okt.pos(content)
words = [word for word, pos in morphs if pos =="Noun"]
print(words)
```

```python
from collections import Counter
stopwords = ['[',']','도리#2417','HABIAD#3138','발컨#5913','발컨','봇','리나','고가도','도리','봇말좀','봇입','고도리','봇욕','의자','기소','당신','우리','규칙','개조','언급']
result = [x for x in words if x not in stopwords and len(x) > 1]
print(result)

c = Counter(result)
print(c.most_common(10))
```

```python
from wordcloud.wordcloud import FONT_PATH
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from os import path

FONT_PATH = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"  

text = " ".join(result)
wordcloud = WordCloud(
    max_font_size=100, 
    max_words=55, 
    background_color= '#FFFBFA', 
    relative_scaling=.5, 
    width = 500,
    height = 300,
    font_path=FONT_PATH).generate(text)

plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("SSM_01.png")
```

```python
nouns = result
nouns
```

```python
import pandas as pd
df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
df_word['count'] = df_word['word'].str.len() 
df_word
```

```python
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
df_word = df_word.groupby('word', as_index=False).agg(n = ('word', 'count')).sort_values('n', ascending = False) 
df_word
```

```python
top20 = df_word.head(20) 
top20
```

```python
font = 'NanumBarunGothic.ttf'
```

```python
dic_word = df_word .set_index('word').to_dict()['n']
dic_word
```

```python
from wordcloud import WordCloud
wc = WordCloud(random_state = 100,
               font_path = font,
               width = 400,
               height = 400,
               background_color = '#fefefe')
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

```python
img_wordcloud.to_file("SSM_01_V2.png")
```
